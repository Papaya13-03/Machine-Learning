```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_url = "./Study_vs_Score_data.csv"
data = pd.read_csv(data_url)
```

```python
def plot_data(data):
    plt.scatter(data['Attendance_Hours'], data['Final_Marks'], color='blue', label='Data Points')
    plt.xlabel('Attendance_Hours')
    plt.ylabel('Final_Marks')
    plt.title('Attendance_Hours vs Final_Marks')
    plt.legend()
    plt.grid()
    plt.show()
plot_data(data)
```

```python
def loss_function(m, b, points):
    total_loss = 0
    data_length = len(points)
    for i in range(data_length):
        point = points.iloc[i]
        _y = m * point['Attendance_Hours'] + b
        y = point['Final_Marks']
        total_loss += (_y - y)**2
    total_loss = 1/len(points) * total_loss
    return total_loss
```

```python
def gradient_descent(m_now, b_now, learning_rate, points):
    grad_m = 0
    grad_b = 0
    data_length = len(points)
    for i in range(data_length):
        point = points.iloc[i]
        x = point['Attendance_Hours']
        y = point['Final_Marks']
        grad_m += -2/data_length*x*(y - (m_now*x + b_now))
        grad_b += -2/data_length*(y - (m_now*x + b_now))
    m_next = m_now - learning_rate * grad_m
    b_next = b_now - learning_rate * grad_b
    return [m_next, b_next]
```

```python
m = 1
b = 1
learning_rate = 0.0001
epochs = 50000

for i in range(epochs):
    m, b = gradient_descent(m, b, learning_rate, data)
    if i%500 == 0:
        loss = loss_function(m, b, data)
        print(f"Epoch: {i}  | Loss: {loss}")

print(f"Final m: {m}, Final b: {b}")
```

```python
plt.scatter(data['Attendance_Hours'], data['Final_Marks'], color='blue', label='Data Points')
x = np.linspace(data['Attendance_Hours'].min(), data['Attendance_Hours'].max(), 100)
y = m * x + b
plt.plot(x, y, color='red', label='Regression Line')
plt.xlabel('Attendance_Hours')
plt.ylabel('Final_Marks')
plt.title('Attendance_Hours vs Final_Marks with Regression Line')
plt.legend()
plt.grid()
plt.show()
```
